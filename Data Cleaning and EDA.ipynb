```python
df['Predictor'] = np.where( ( (df['All_Star_apps'] >= 1) ) |
                          ( (df['All_Star_apps'] == 0) & (df['NBA_PER'] > 13.8) & (df['NBA_Tenure'] > 2) ) |
                          ( (df['All_Star_apps'] == 0) & (df['NBA_PER'] > 9.7) & (df['NBA_PER'] < 13.8) & (df['NBA_Tenure'] > 2) ), 3, 2, 1)
```

```python
#Create the target variable 'tier'
# create a list of our conditions
conditions = [
    (df['All_Star_apps'] >= 1),
    (df['All_Star_apps'] == 0) & (df['NBA_PER'] >= 13.8) & (df['NBA_Tenure'] > 2),
    (df['All_Star_apps'] == 0) & (df['NBA_PER'] > 9.7) & (df['NBA_PER'] < 13.8) & (df['NBA_Tenure'] > 2),
    (df['All_Star_apps'] == 0) & (df['NBA_PER'] <= 9.7) & (df['NBA_Tenure'] > 2),
    (df['All_Star_apps'] == 0) & (df['NBA_Tenure'] <= 2)
    ]

# create a list of the values we want to assign for each condition
values = [3, 2, 1, 0, 0]

# create a new column and use np.select to assign values to it using our lists as arguments
df['Tier'] = np.select(conditions, values)

# display updated DataFrame
df.head(25)
```
